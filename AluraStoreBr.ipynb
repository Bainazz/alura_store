### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

```python
# outros imports
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#1. Análise do faturamento


```python
# teste
loja1['Preço']
```

```python
# teste
loja1.info()
```

```python
# calculos
faturamento_loja1 = (loja1['Preço'] + loja1['Frete']).sum().round(2)
faturamento_loja2 = (loja2['Preço'] + loja2['Frete']).sum().round(2)
faturamento_loja3 = (loja3['Preço'] + loja3['Frete']).sum().round(2)
faturamento_loja4 = (loja4['Preço'] + loja4['Frete']).sum().round(2)

# saídas
print(f'Faturamento total da loja 1: R$ {faturamento_loja1}')
print(f'Faturamento total da loja 2: R$ {faturamento_loja2}')
print(f'Faturamento total da loja 3: R$ {faturamento_loja3}')
print(f'Faturamento total da loja 4: R$ {faturamento_loja4}')

# criacao da lista com todos os faturamentos
resultado_faturamento = [faturamento_loja1, faturamento_loja2,faturamento_loja3, faturamento_loja4]
```

# 2. Vendas por Categoria


```python
# testes
categoria = loja1['Categoria do Produto'].unique()
```

```python
# calculos
contagem_vendas_loja1 = loja1['Categoria do Produto'].value_counts()
contagem_vendas_loja2 = loja2['Categoria do Produto'].value_counts()
contagem_vendas_loja3 = loja3['Categoria do Produto'].value_counts()
contagem_vendas_loja4 = loja4['Categoria do Produto'].value_counts()

# saídas
print(f'Vendas da loja 1: \n{contagem_vendas_loja1}')
print()
print(f'Vendas da loja 2: \n{contagem_vendas_loja2}')
print()
print(f'Vendas da loja 3: \n{contagem_vendas_loja3}')
print()
print(f'Vendas da loja 4: \n{contagem_vendas_loja4}')
print()
```

# 3. Média de Avaliação das Lojas

```python
# calculos
contagem_avaliacao_loja1 = loja1['Avaliação da compra'].mean().round(2)
contagem_avaliacao_loja2 = loja2['Avaliação da compra'].mean().round(2)
contagem_avaliacao_loja3 = loja3['Avaliação da compra'].mean().round(2)
contagem_avaliacao_loja4 = loja4['Avaliação da compra'].mean().round(2)

# saídas
print(f'Média de avaliações da loja 1: {contagem_avaliacao_loja1} ☆')
print(f'Média de avaliações da loja 2: {contagem_avaliacao_loja2} ☆')
print(f'Média de avaliações da loja 3: {contagem_avaliacao_loja3} ☆')
print(f'Média de avaliações da loja 4: {contagem_avaliacao_loja4:.2f} ☆')

# criacao da lista com todos os fretes
resultado_avaliacao = [contagem_avaliacao_loja1, contagem_avaliacao_loja2,contagem_avaliacao_loja3, contagem_avaliacao_loja4]
```

# 4. Produtos Mais e Menos Vendidos

```python
# calculos
maior_venda_loja1 = loja1['Produto'].value_counts().idxmax(), loja1['Produto'].value_counts().max()
maior_venda_loja2 = loja2['Produto'].value_counts().idxmax(), loja2['Produto'].value_counts().max()
maior_venda_loja3 = loja3['Produto'].value_counts().idxmax(), loja3['Produto'].value_counts().max()
maior_venda_loja4 = loja4['Produto'].value_counts().idxmax(), loja4['Produto'].value_counts().max()
menor_venda_loja1 = loja1['Produto'].value_counts().idxmin(), loja1['Produto'].value_counts().min()
menor_venda_loja2 = loja2['Produto'].value_counts().idxmin(), loja2['Produto'].value_counts().min()
menor_venda_loja3 = loja3['Produto'].value_counts().idxmin(), loja3['Produto'].value_counts().min()
menor_venda_loja4 = loja4['Produto'].value_counts().idxmin(), loja4['Produto'].value_counts().min()

# saídas
print(f'Produto mais vendido da loja 1: {maior_venda_loja1[0]}, {maior_venda_loja1[1]} unidades vendidas')
print(f'Produto menos vendido da loja 1: {menor_venda_loja1[0]}, {menor_venda_loja1[1]} unidades vendidas')
print()
print(f'Produto mais vendido da loja 2: {maior_venda_loja2[0]}, {maior_venda_loja2[1]} unidades vendidas')
print(f'Produto menos vendido da loja 2: {menor_venda_loja2[0]}, {menor_venda_loja2[1]} unidades vendidas')
print()
print(f'Produto mais vendido da loja 3: {maior_venda_loja3[0]}, {maior_venda_loja3[1]} unidades vendidas')
print(f'Produto menos vendido da loja 3: {menor_venda_loja3[0]}, {menor_venda_loja3[1]} unidades vendidas')
print()
print(f'Produto mais vendido da loja 4: {maior_venda_loja4[0]}, {maior_venda_loja4[1]} unidades vendidas')
print(f'Produto menos vendido da loja 4: {menor_venda_loja4[0]}, {menor_venda_loja4[1]} unidades vendidas')
print()
```

# 5. Frete Médio por Loja


```python
#calculos
frete_loja1 = loja1['Frete'].mean().round(2)
frete_loja2 = loja2['Frete'].mean().round(2)
frete_loja3 = loja3['Frete'].mean().round(2)
frete_loja4 = loja4['Frete'].mean().round(2)

#saída
print(f'Média de frete da loja 1: R$ {frete_loja1}')
print(f'Média de frete da loja 2: R$ {frete_loja2}')
print(f'Média de frete da loja 3: R$ {frete_loja3}')
print(f'Média de frete da loja 4: R$ {frete_loja4}')

# criacao da lista com todos os fretes
resultado_frete = [frete_loja1, frete_loja2,frete_loja3, frete_loja4]
```

# 6. Gráficos

## Faturamento

```python
# grafico pizza - faturamento total
nome_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# funcao dos valores abaixo da porcentagem
def formatar_autopct(valores):
    def func(pct):
        total = sum(valores)
        valor = pct * total / 100
        return f'{pct:.1f}%\nR$ {valor:,.2f}'
    return func

# o grafico em si
plt.figure(figsize=(6, 6))
plt.pie(resultado_faturamento,
        labels= nome_lojas,
        autopct=formatar_autopct(resultado_faturamento),
        startangle=90,
        colors=['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3'])
plt.title('Faturamento por Loja')
plt.axis('equal')
plt.show()
```

## Avaliações


```python
# avaliacoes loja 1

# conversao da data
loja1['Data da Compra'] = pd.to_datetime(loja1['Data da Compra'], dayfirst=True)

# coluna com ano e mês
loja1['AnoMes'] = loja1['Data da Compra'].dt.to_period('M')

# agrupa mês e calcula média
media_mensal = loja1.groupby('AnoMes')['Avaliação da compra'].mean()

# indice pra datetime
media_mensal.index = media_mensal.index.to_timestamp()

plt.figure(figsize=(12, 6))
plt.plot(media_mensal.index, media_mensal.values, marker='o', linestyle='-')
plt.title('Média mensal das avaliações - Loja 1')
plt.xlabel('Meses')
plt.ylabel('Avaliação média')
plt.ylim(3, 5)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Produtos mais e menos vendidos


```python
# produtos mais vendidos - por loja
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

produtos_mais_vendidos = [maior_venda_loja1[1], maior_venda_loja2[1], maior_venda_loja3[1], maior_venda_loja4[1]]
nome_produtos = [maior_venda_loja1[0], maior_venda_loja2[0], maior_venda_loja3[0], maior_venda_loja4[0]]

plt.figure(figsize=(10,6))
bars = plt.bar(lojas, produtos_mais_vendidos)
plt.title('Produto mais vendido por loja')
plt.ylabel('Unidades vendidas')
plt.ylim(0, max(produtos_mais_vendidos) + 10)

for bar, produto in zip(bars, nome_produtos):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             produto, ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# produtos menos vendidos - por loja
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

produtos_menos_vendidos = [menor_venda_loja1[1], menor_venda_loja2[1], menor_venda_loja3[1], menor_venda_loja4[1]]
nome_produtos = [menor_venda_loja1[0], menor_venda_loja2[0], menor_venda_loja3[0], menor_venda_loja4[0]]

plt.figure(figsize=(10,6))
bars = plt.bar(lojas, produtos_menos_vendidos)
plt.title('Produto menos vendido por loja')
plt.ylabel('Unidades vendidas')
plt.ylim(0, max(produtos_menos_vendidos) + 10)

for bar, produto in zip(bars, nome_produtos):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             produto, ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

## Frete


```python
# grafico de barras - preço do frete

plt.figure(figsize=(8, 5))
plt.bar(lojas, resultado_frete)
plt.title('Média do valor de frete por loja')
plt.xlabel('Loja')
plt.ylabel('Frete médio (R$)')
plt.ylim(0, max(resultado_frete) + 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)

for i, valor in enumerate(resultado_frete):
    plt.text(i, valor + 0.5, f'R$ {valor:.2f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
